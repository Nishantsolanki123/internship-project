```python
# Importing essential libraries for data manipulation, visualization, and modeling
import pandas as pd
```

```python
# Loading the dataset from the source to analyze and preprocess
df = pd.read_csv('/content/Student Performance Predictor for EduQuest Coaching.csv')
```

```python
df.head()
```

```python
# Displaying basic statistics and structure of the data to understand its properties
df.info()
```

```python
# Handling missing values, encoding categorical variables, and scaling features for model compatibility
df=df.dropna()
```

```python
df.describe(include ='all')
```

```python
# Checking categorical data
df['gender'].value_counts()
```

```python
df['gender'] = df['gender'].replace({'Female':1, 'Male':0})
```

```python
# Checking categorical data
df['parental_education'].value_counts()
```

```python
# Ordinal encoding
df['parental_education'] = df['parental_education'].replace({'Diploma':0,'High School':1,
                                  'Graduate':2,'Postgraduate':3})
```

```python
df['internet_access'].value_counts()
```

```python
df['internet_access'] = df['internet_access'].replace({'Yes':1, 'No':0})
```

```python
df['tutor_support'].value_counts()
```

```python
df['tutor_support'] = df['tutor_support'].replace({'Yes':1, 'No':0})
```

```python
df['extra_curricular_involvement'].value_counts()
```

```python
df['extra_curricular_involvement'] = df['extra_curricular_involvement'].replace({'Low':0,'Moderate':1,'High':2})
```

```python
df.info()
```

```python
# Define y and X
df.columns
```

```python
y = df['final_exam_score']
X = df[['gender', 'age', 'parental_education', 'family_income',
       'internet_access', 'previous_exam_score', 'attendance_rate',
       'homework_completion_rate', 'class_participation_score',
       'number_of_absences', 'extra_curricular_involvement',
       'learning_hours_per_week', 'tutor_support']]
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2529)
```

```python
# Select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Train model
model.fit(X_train,y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
# Evaluate
# Summarizing findings, model accuracy, and potential next steps or improvements
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # ideal line
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

```python
import pickle

# Save model to file
with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
import gradio as gr
import numpy as np
import pickle

# Load your trained model
model = pickle.load(open("model.pkl", "rb"))

def predict_fn(gender, age, parental_education, family_income,
               internet_access, previous_exam_score, attendance_rate,
               homework_completion_rate, class_participation_score,
               number_of_absences, extra_curricular_involvement,
               learning_hours_per_week, tutor_support):

    input_data = np.array([[gender, age, parental_education, family_income,
                            internet_access, previous_exam_score, attendance_rate,
                            homework_completion_rate, class_participation_score,
                            number_of_absences, extra_curricular_involvement,
                            learning_hours_per_week, tutor_support]])

    prediction = model.predict(input_data)[0]
    prediction = np.clip(prediction, 0, 100)  #  ensure the score stays within 0100
    return f"Predicted Score: {prediction:.2f}"

# Gradio UI
gr.Interface(
    fn=predict_fn,
    inputs=[
        gr.Number(label="Gender (0=Male, 1=Female)"),
        gr.Number(label="Age"),
        gr.Number(label="Parental Education (Diploma=0, High School=1, Graduate=2, Postgraduate=3)"),
        gr.Number(label="Family Income"),
        gr.Number(label="Internet Access (0=No, 1=Yes)"),
        gr.Number(label="Previous Exam Score"),
        gr.Number(label="Attendance Rate (%)"),
        gr.Number(label="Homework Completion Rate (%)"),
        gr.Number(label="Class Participation Score (010)"),
        gr.Number(label="Number of Absences"),
        gr.Number(label="Extra Curricular Involvement (0=No, 1=Yes)"),
        gr.Number(label="Learning Hours Per Week"),
        gr.Number(label="Tutor Support (0=No, 1=Yes)")
    ],
    outputs=gr.Text(label="Predicted Final Exam Score"),
    title="Student Performance Predictor",
    description="Enter the data of student to predict student performance.",
    live=True
).launch()
```
